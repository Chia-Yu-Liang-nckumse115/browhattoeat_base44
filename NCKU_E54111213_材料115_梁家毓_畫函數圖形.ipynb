```python
# 讀入標準套件: numpy, pandas, matplotlib
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

1. 主題: 鋼鐵人的反應爐(中間圓圓那個)
2. 嘗試 "classic"風格
3. 先畫外圈的大圓
4. 移除角度範圍第三個elements的部分
5. 內接一個倒三角形
6. 在三角形中間在畫一個小三角形
7. 設定縮小比例0.5倍
8. 將小三角形三個角的地方和大三角型的三個角相連(用直線)
9. 將背景調黑
10. 將線條調白(在plt.plot 括弧後面加一個 color = "想要的顏色就可以了")
11. 移除座標軸
12. 小發現:原本我的數字和文字之間沒有空格所以字都沒有換行很難看(要記得空格)


```python
# 設定背景為黑色 dark_background 因為我想要反應爐是白色的(亮的)
plt.style.use('dark_background')

# 設定圓形的半徑
r = 5

# 設定角度的範圍
theta = np.linspace(0, 2*np.pi)

# 寫出 x 和 y 的座標
x_circle = r * np.cos(theta)
y_circle = r * np.sin(theta)

# 畫出圓形
plt.plot(x_circle, y_circle, color = "white")

# 計算內接正三角形的三個頂點座標(prompt: 園內接倒三角形)
# 倒三角形，所以第一個點在下方
x_triangle = [r * np.cos(3 * np.pi / 2), r * np.cos(np.pi / 6), r * np.cos(5 * np.pi / 6), r * np.cos(3 * np.pi / 2)]
y_triangle = [r * np.sin(3 * np.pi / 2), r * np.sin(np.pi / 6), r * np.sin(5 * np.pi / 6), r * np.sin(3 * np.pi / 2)]

# 畫出三角形
plt.plot(x_triangle, y_triangle, color = "white")

# 計算縮小後的小三角形的頂點座標(prompt: 在三角形內在畫一個等比例縮小的三角形，小三角形和大三角形三個角需要相連)
scale_factor = 0.5 #小三角形的縮小倍率
x_triangle_small = [x_triangle[i] * scale_factor for i in range(3)]
y_triangle_small = [y_triangle[i] * scale_factor for i in range(3)]

# 為了封閉小三角形，將第一個點加回末尾
x_triangle_small.append(x_triangle_small[0])
y_triangle_small.append(y_triangle_small[0])

# 畫出縮小後的小三角形
plt.plot(x_triangle_small, y_triangle_small, color = "white")

# 連接大三角形和小三角形的對應頂點
for i in range(3):
    plt.plot([x_triangle[i], x_triangle_small[i]], [y_triangle[i], y_triangle_small[i]], color='white') # 將線條顏色改為白色

# 設定座標軸的比例，讓圓形看起來是圓的(沒有設定座標的時候是橢圓形的)
plt.axis('equal')


# 移除座標軸
plt.axis('off')


# 顯示圖形
plt.show()
```
